Encoding Wind Direction Data

```python
def convert_wd(column):
  for row in range(column.shape[0]):
    if column[row] == 'N': 
      column[row] = 0
    elif column[row] == 'NNE': 
      column[row] = 22.5
    elif column[row] == 'NE': 
      column[row] = 45
    elif column[row] == 'ENE': 
      column[row] = 67.5
    elif column[row] == 'E': 
      column[row] = 90
    elif column[row] == 'ESE': 
      column[row] = 112.5
    elif column[row] == 'SE': 
      column[row] = 135
    elif column[row] == 'SSE': 
      column[row] = 157.5
    elif column[row] == 'S': 
      column[row] = 180
    elif column[row] == 'SSW': 
      column[row] = 202.5
    elif column[row] == 'SW': 
      column[row] = 225
    elif column[row] == 'WSW': 
      column[row] = 247.5
    elif column[row] == 'W': 
      column[row] = 270
    elif column[row] == 'WNW': 
      column[row] = 292.5
    elif column[row] == 'NW': 
      column[row] = 315
    elif column[row] == 'NNW':
      column[row] = 337.5
    elif column[row] == 'N': 
      column[row] = 360
  return column

df0['wd'] = convert_wd(df0['wd'])
df1['wd'] = convert_wd(df1['wd'])
df2['wd'] = convert_wd(df2['wd'])
df3['wd'] = convert_wd(df3['wd'])
df4['wd'] = convert_wd(df4['wd'])
df5['wd'] = convert_wd(df5['wd'])
df6['wd'] = convert_wd(df6['wd'])
df7['wd'] = convert_wd(df7['wd'])
df8['wd'] = convert_wd(df8['wd'])
df9['wd'] = convert_wd(df9['wd'])
df10['wd'] = convert_wd(df10['wd'])
df11['wd'] = convert_wd(df11['wd'])
```

Create Early Stopping Class

```python
class EarlyStopping:
  def __init__(self, patience=5, min_delta=0):
    self.patience = patience
    self.min_delta = min_delta
    self.counter = 0
    self.early_stop = False

  def __call__(self, train_loss, test_loss):
    if (test_loss - train_loss) > self.min_delta:
      self.counter +=1
      if self.counter >= self.patience:  
        self.early_stop = True

early_stopping = EarlyStopping(patience=50, min_delta=0.001)
```

Convert Features to Supervized Data and Split Train and Test Data

```python
import torch

train_DataFrame, test_DataFrame = feature.loc[0:28051], feature.loc[28052:]

def convert_to_supervised(feature, batch):
  feature_torch = torch.tensor(feature.values)
  X, Y = [], []
  for i in range(batch, feature_torch.size()[0]-1):
    X.append(feature_torch[i-batch:i, :])
    Y.append(feature_torch[i+1, -1])
  return X, Y

class my_data():
  def __init__(self, data, label):
    self.data = data
    self.label = label

  def __len__(self):
    return len(self.data)

  def __getitem__(self, index):
    return self.data[index], self.label[index]
```

Import Daily and Weekly Data in DataLoader

```python
train_data_day, train_label_day = convert_to_supervised(train_DataFrame, 24)
test_data_day, test_label_day = convert_to_supervised(test_DataFrame, 24)

train_torch_day = my_data(train_data_day, train_label_day)
test_torch_day = my_data(test_data_day, test_label_day)

train_day = torch.utils.data.DataLoader(train_torch_day, batch_size=512, shuffle=True)
test_day = torch.utils.data.DataLoader(test_torch_day, batch_size=256, shuffle=True)
```

Create CNN-LSTM Model

```python
import torch.nn as nn

class GetLSTMOutput(nn.Module):
    def forward(self, x):
        out, _ = x
        return out

# Casual_Padding = (kernel_size-1)*dilation

CNN_LSTM_day = nn.Sequential(
    nn.Conv1d(24, 64, kernel_size=3, padding=2, dilation=1),
    nn.ReLU(),
    nn.BatchNorm1d(64),
    nn.Dropout(0.1),

    nn.Conv1d(64, 64, kernel_size=3, padding=2, dilation=1),
    nn.ReLU(),
    nn.BatchNorm1d(64),
    nn.Dropout(0.1),

    nn.Conv1d(64, 32, kernel_size=3, padding=2, dilation=1),
    nn.ReLU(),
    nn.Dropout(0.1),

    nn.MaxPool1d(kernel_size=3, padding=1, dilation=1),

    nn.LSTM(9, 64, num_layers=100, dropout=0.2, batch_first=True),
    GetLSTMOutput(),

    nn.LSTM(64, 32, num_layers=50, dropout=0.3, batch_first=True),
    GetLSTMOutput(),

    nn.Linear(32, 1),
    nn.ReLU()
)
```

**Daily:** Create Optimizer, Scheduler, and Loss Function

```python
optimizer = torch.optim.Adam(CNN_LSTM_day.parameters(), lr = 0.001)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=32, gamma=0.0001)

loss_function = nn.CrossEntropyLoss()

softmax = nn.Softmax(dim=1)
```

**Daily:** Create Device and Add Loss Function and Model to Device

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
CNN_LSTM_day.to(device)
loss_function.to(device)
```

**Daily:** Train Model

```python
epochs_number = 200
train_loss_list, test_loss_list = [], []

for epoch in range(epochs_number):
  train_loss, test_loss = [], []
  CNN_LSTM_day.train()
  for data, pred in train_day:
    data = data.to(device=device, dtype=torch.float)
    pred = pred.to(device=device, dtype=torch.int64)
    optimizer.zero_grad()
    output = CNN_LSTM_day(data)
    loss = loss_function(output, pred.unsqueeze(1).clone().detach().to(device))
    loss.backward()
    train_loss.append(loss.item())
    optimizer.step()
  scheduler.step()
  
  with torch.no_grad():
    CNN_LSTM_day.eval()
    for data, pred in test_day:
      data = data.to(device=device, dtype=torch.float)
      pred = pred.to(device=device, dtype=torch.int64)
      outputs = CNN_LSTM_day(data)
      outputs = softmax(outputs)
      los = loss_function(outputs, pred.unsqueeze(1))
      test_loss.append(los.item())

    train_loss_list.append(sum(train_loss)/len(train_loss))
    test_loss_list.append(sum(test_loss)/len(test_loss))

    if(test_loss_list[-1]<=min(test_loss_list)):
      print(f'in Epoch {epoch}'+' Model Saved!')
      torch.save(CNN_LSTM_day, 'Best_Daily_Model.pth')

    early_stopping(sum(train_loss)/len(train_loss), sum(test_loss)/len(test_loss))
    print(f'Epoch {epoch} : Train Loss {sum(train_loss)/len(train_loss)}   Test Loss {sum(test_loss)/len(test_loss)}')

  if early_stopping.early_stop:
    print(f'in Epoch {epoch} Train Break!')
    break
```

**Daily:** Test Model

```python
CNN_LSTM_day.eval() 

datas = torch.Tensor().to('cuda')
preds = torch.Tensor().to('cuda')

with torch.no_grad():
  for data, pred in test_day:
    data = data.to(device=device, dtype=torch.float)
    pred = pred.to(device=device, dtype=torch.int64)
    outputs = CNN_LSTM_day(data)
    outputs = softmax(outputs)
    los = loss_function(outputs, pred.unsqueeze(1))
    datas = torch.cat((datas, pred))
    preds = torch.cat((preds, outputs))
```

**Daily:** Plot Train Loss and Test Loss

```python
import matplotlib.pyplot as plt
plt.plot(train_loss_list)
plt.plot(test_loss_list)
plt.legend(['train loss', 'test loss'])
plt.show()
```

**Daily:** MAE, RMSE, and R2 Coeffcient 

```python
n = len(datas)

MAE = torch.sum(torch.abs(datas - preds[:, 0].squeeze())) / n
print(f'Mean Absolute Error = {MAE}')

RMSE = torch.sqrt(torch.sum(torch.pow(datas - preds[:, 0].squeeze(), 2)) / n)
print(f'Root Mean Square Error = {RMSE}')

R2 = torch.sum(torch.pow(datas - preds[:, 0].squeeze(), 2)) / torch.sum(torch.pow(datas - torch.mean(preds[:, 0].squeeze()), 2))
print(f'Mean Absolute Error = {R2}')
```
