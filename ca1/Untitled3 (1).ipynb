**B.** Merge Movies and Ratings Data

```python
# B
merged_data = movies_data.merge(ratings_data, how = 'inner', on = 'movieId')
merged_data
```

**C.** Delete Extra Columns

```python
# C
merged_data.drop('title', inplace = True, axis = 1)
merged_data.drop('genres', inplace = True, axis = 1)
merged_data.drop('timestamp', inplace = True, axis = 1)
merged_data
```

**E.1.** Normalize Users Ratins

```python
# E.1
merged_data['rating'] = merged_data['rating'].transform(lambda x: (x - x.mean()) / x.std())
merged_data
```

**D.** Group Data with GroupBy Method

```python
# D
data = merged_data.groupby('userId', sort = True)
data.first()
# merged_data.groupby('userId')['movieId'].nunique()
```

**E.2.** Create train_X List

```python
# E.2
train_X = data['rating']
train_X = train_X.apply(pd.DataFrame)
train_X = train_X.to_numpy()
```

**F.** Create Neurons

```python
# F
hidden_nueron_number = len(data)
visible_nueron_number = np.shape(movies_data)[0]

class Restricted_Boltzmann_Machine():
  def __init__(self, hidden_node, visible_node):
    self.weights = np.random.normal(0, 1, [hidden_node, visible_node])
    self.bias_hidden_give_visible = np.random.normal(0, 1, [1, hidden_node])
    self.bias_visible_give_hidden = np.random.normal(0, 1, [1, visible_node])
    print(np.shape(self.bias_visible_give_hidden))

  def hidden_layer(self, x):
    weights_hidden_node = np.dot(x, self.weights)
    # print(np.shape(weights_hidden_node))
    activation_function = weights_hidden_node + self.bias_visible_give_hidden*np.ones([np.shape(weights_hidden_node)[0], np.shape(weights_hidden_node)[1]])
    activation_function = torch.FloatTensor(activation_function)
    return torch.sigmoid(activation_function)

  def visible_layer(self, x):
    weights_visible_node = np.dot(x, np.transpose(self.weights))
    activation_function = weights_visible_node + self.bias_hidden_give_visible*np.ones([np.shape(weights_visible_node)[0], np.shape(weights_visible_node)[1]])
    activation_function = torch.FloatTensor(activation_function)
    return torch.sigmoid(activation_function)

  def train(self, v0, vk, ph0, phk, lr):
    self.weights += lr*(np.dot(np.transpose(v0), ph0) - np.dot(np.transpose(vk), phk))
    self.bias_hidden_give_visible += lr*(np.sum((ph0 - phk), axis = 0))
    self.bias_visible_give_hidden += lr*(np.sum((v0 - vk), axis = 0))
```

```python
epochs = 20
batch = 100
users_number = np.shape(data)[0]
RBM = Restricted_Boltzmann_Machine(hidden_nueron_number, visible_nueron_number)

for epoch in range(1, epochs + 1):
    train_loss = 0
    s = 0.
    for id_user in range(0, users_number - batch, batch):
        vk = train_X[id_user:id_user+batch]
        # print(id_user,type(vk))
        v0 = train_X[id_user:id_user+batch]
        ph0 = RBM.hidden_layer(v0)
        for k in range(10):
            hk = RBM.hidden_layer(vk)
            # print(type(v0))
            # print(type(vk))
            # print(type(ph0))
            # print(type(hk))
            # print(k,type(vk))
            vk = RBM.visible_layer(hk)
            vk = vk.detach().numpy()
            # print(k,type(vk))
            # vk = vk.to_numpy()
            vk[v0<0] = v0[v0<0]
        vk=torch.from_numpy(vk)
        # print(type(vk))
        phk = RBM.hidden_layer(vk)
        # vk = vk.detach().numpy()
        RBM.train(v0, vk, ph0, phk, 0.001)
        train_loss += torch.mean(torch.abs(v0[v0>=0] - vk[v0>=0]))
        s += 1.
        # vk = vk.detach().numpy()
    print('epoch: '+str(epoch)+' loss: '+str(train_loss/s))
```
